1. Crear una clase Punto con los siguientes atributos privados: x e y
* **constructor** (x e y por default 0)
* **getters**
* **\_\_str\_\_()**: representación textual con formato: (x, y)
* **get_nante()**: cuadrante en el cual se encuentra el punto. Usar enumerado Cuadrante: I, II, III, IV
* **distancia(Punto)**: distancia entre dos puntos. Usar Pitágoras: $d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$

```python
from enum import Enum

class Cuadrante(Enum):
    I = 1
    II = 2
    III = 3
    IV = 4
    SIN = 0
```

```python
import math

class Punto:

    def __init__(self, x=0, y=0):
        self.__x = x
        self.__y = y

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def __str__(self):
        return f'({self.__x}, {self.__y})'

    def __repr__(self):
        return self.__str__()

    def get_nante(self):
        cuad = Cuadrante.SIN
        if self.__x > 0 and self.__y > 0:
            cuad = Cuadrante.I
        if self.__x < 0 and self.__y > 0:
            cuad = Cuadrante.II
        if self.__x < 0 and self.__y < 0:
            cuad = Cuadrante.III
        if self.__x > 0 and self.__y < 0:
            cuad = Cuadrante.IV
        return cuad.name

    def distancia(self, punto):
        return math.sqrt(math.pow(punto.get_x() - self.get_x(), 2) + math.pow(punto.get_y() - self.get_y(), 2))
```

```python
p1 = Punto()
p2 = Punto(3, 4)
print(p1, p2)
print(p1.get_nante(), p2.get_nante())
print(p1.distancia(p2))
```

2. Crear una clase Punto3D que extienda a la clase Punto con el atributo privado y propio z
* **constructor** (x, y, z por default 0)
* **get_z()**
* **\_\_str\_\_()**: representación textual con formato: (x, y, z)

```python
class Punto3D(Punto):

    def __init__(self, x=0, y=0, z=0):
        super().__init__(x, y)
        self.__z = z

    def get_z(self):
        return self.__z

    def __str__(self):
        return f'({self.get_x()}, {self.get_y()}, {self.get_z()})'

    def __repr__(self):
        return self.__str__()
```

3. Describa brevemente el concepto de **sobreescritura**

* Es un concepto de la herencia que permite a una clase hija (subclase) redefinir un método que ya existe en su clase padre (superclase)
* Esto permite que la subclase tenga un comportamiento distinto o adicional para ese método
* Para lograrlo, se crea un método con el mismo nombre en la clase hija, haciendo que al llamar el método en una instancia de la subclase se ejecute la nueva implementación

1. Crear una clase Plano con los siguientes atributos: x_name, y_name y una lista de objetos punto
* **constructor** (x_name e y_name por default 'ejeX' y 'ejeY' respectivamente y la lista se crea vacia)
* **agregar_punto(Punto)**. Lanzar excepción si se quiere agregar un Punto3D en un plano. Tratarla creando un punto con valores de x e y del objeto que entro por parámetro
* **sumarizar()**: cantidad de puntos por cuadrante. Usar diccionario
* **descargar()**: escribir en archivo puntos.csv los puntos de la lista: x;y;cuadrante

```python
class Plano:

    def __init__(self, x_name='ejeX', y_name='ejeY'):
        self.__x_name = x_name
        self.__y_name = y_name
        self.__lista = []

    def agregar_punto(self, punto):
        try:
            if type(punto) == Punto3D:
                raise Exception
        except Exception:
            punto = Punto(punto.get_x(), punto.get_y())
        self.__lista.append(punto)

    def sumarizar(self):
        mapa = {'I': 0, 'II': 0, 'III':0, 'IV':0, 'SIN': 0}
        for punto in self.__lista:
            mapa[punto.get_nante()] += 1
        return mapa

    def descargar(self):
        fw = open('data/puntos.csv', 'w')
        fw.write('x;y;cuadrante \n')
        for punto in self.__lista:
            fw.write(f'{punto.get_x()};{punto.get_y()};{punto.get_nante()} \n')
        fw.close()
    
    def __str__(self):
        return f'{self.__x_name} y {self.__y_name}: {self.__lista}'
```

```python
plano = Plano()
plano.agregar_punto(Punto(1, 2))
print(plano)
plano.agregar_punto(Punto3D(3, 4, 5))
print(plano)
print(plano.sumarizar())
plano.descargar()
```
