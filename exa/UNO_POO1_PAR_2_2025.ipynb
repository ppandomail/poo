1. Crear una clase llamada **Cancion** con los siguientes atributos: titulo, artista, duracion_minutos, reproducciones y genero. Debe incluir:
* **Constructor**
* Método **\_\_str__()** que devuelva la representación textual del objeto con el siguiente formato: "Flowers - Miley Cyrus (3.2 min, 2.500.000 reproducciones)"
* Método **es_popular()** que devuelva True si tiene más de 5.000.000 reproducciones.

```python
class Cancion:

    def __init__(self, titulo, artista, duracion_minutos, reproducciones, genero):
        self.__titulo = titulo
        self.__artista = artista
        self.__duracion_minutos = duracion_minutos
        self.__reproducciones = reproducciones
        self.__genero = genero

    def __str__(self):
        return f'{self.__titulo} - {self.__artista} ({self.__duracion_minutos} min, {self.__reproducciones} reproducciones)'

    def __repr__(self):
        return self.__str__()
    
    def es_popular(self):
        return self.__reproducciones > 5000000

    def get_genero(self):
        return self.__genero
```

2. Implementar una clase llamada **Playlist** con un método **cargar_playlist(ruta_archivo)** que lea de un archivo con extensión csv que contiene canciones y almacene en una lista objetos **Canción**. Cada línea del archivo tiene la siguiente estructura:

| titulo    | artista     | duracion_minutos | reproducciones | genero    |
| --        | --          | --               | --             | --        |
| Flowers   | Miley Cyrus | 3.20             | 2500000        | Pop       |
| Despacito | Luis Fonsi  | 4.45             | 8000000        | Reggaeton |
| Unholy    | Sam Smith   | 3.15             | N/A            | Pop       |

Debe manejar las siguientes excepciones:
* **FileNotFoundError**: mostrar "Archivo no encontrado".
* **ValueError**: si alguna duración o cantidad de reproducciones no se puede convertir, mostrar "No es un número"
* Cualquier otra excepción debe capturarse y mostar "Error!!".

```python
import csv
class Playlist:

    def __init__(self):
        self.__canciones = []

    def cargar_playlist(self, ruta_archivo):
        try:
            with open(ruta_archivo, "r", encoding="utf-8") as f:
                datos = csv.DictReader(f, delimiter=';')
                for reg in datos:
                    self.__canciones.append(Cancion(reg["titulo"], reg["artista"], float(reg["duracion_minutos"]), int(reg["reproducciones"]), reg["genero"]))
        except FileNotFoundError:
            print('Archivo no encontrado')
        except ValueError:
            print('No es un número')
        except Exception as e:
            print(e)

    def canciones_por_genero(self):
        mapa = {}
        for cancion in self.__canciones:
            if cancion.get_genero() not in mapa:
                mapa[cancion.get_genero()] = []
            mapa[cancion.get_genero()].append(cancion)
        return mapa
```

```python
p = Playlist()
p.cargar_playlist('data/playlist.csv')
p.canciones_por_genero()
```
