1. Una pequeña tienda online quiere digitalizar la gestión de sus productos. Actualmente vende distintos tipos de artículos y necesita un sistema que permita representar diferentes productos que se venden en el catálogo, cada uno con su nombre, precio y cantidad en stock (por default 0). No se permite tener instancias del producto. Algunos productos son prendas de vestir y deben incluir información sobre su talle. Otros productos son electrónicos y deben indicar cuántos meses de garantía tienen. Además, la tienda desea poder mostrar en pantalla la información de cada producto sin importar de qué tipo sea. La super clase debe llevar registro de la cantidad de objetos instanciados y mediante un método retornar su valor.

```python
from abc import abstractmethod, ABC

class Producto(ABC):

    __contador = 0
    
    def __init__(self, nombre, precio, cantidad=0):
        self.__nombre = nombre
        self.__precio = precio
        self.__cantidad = cantidad
        Producto.__contador += 1

    @abstractmethod
    def info(self):
        return self.__nombre + ', ' + str(self.__precio) + ', ' + str(self.__cantidad) + ', '

    def get_precio(self):
        return self.__precio

    @staticmethod
    def get_contador():
        return Producto.__contador

class Prenda(Producto):

    def __init__(self, nombre, precio, talle, cantidad=0):
        super().__init__(nombre, precio, cantidad)
        self.__talle = talle

    def info(self):
        return super().info() + str(self.__talle)

class Electronico(Producto):

    def __init__(self, nombre, precio, meses, cantidad=0):
        super().__init__(nombre, precio, cantidad)
        self.__meses = meses

    def info(self):
        return super().info() + str(self.__meses)
        
p = Prenda('camisa', 100, 40, 30)
e = Electronico('pc', 800, 12, 9)
print(p.info())
print(e.info())
print(Producto.get_contador())        
```

2. Implemente un método que, dada una lista de productos y un monto a gastar, permita conocer cuantos productos pueden comprarse sin superarlo. Tener en cuenta que los productos se encuentran ordenados por prioridad de compra. Realizar un test que verifique que el método funcione correctamente.

```python
def get_cant_prod_comprar(lista, monto):
    if not lista:
        return 0
    cont = 0
    acum = monto
    for prod in lista:
        if prod.get_precio() <= acum:
            cont += 1
            acum -= prod.get_precio()
        if acum == 0:
            break
    return cont    
```

```python
import unittest

class TestCompra(unittest.TestCase):

    def test_cant_prod_comprar(self):
        lista = [Prenda('', 10, 0, 0), Prenda('', 10, 0, 0), Prenda('', 15, 0, 0), Prenda('', 10, 0, 0)]
        self.assertEqual(get_cant_prod_comprar(lista, 40), 3)

unittest.main(defaultTest='TestCompra', argv=[''], verbosity=2, exit=False)
```
