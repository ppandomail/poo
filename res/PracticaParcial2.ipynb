### Ejercicio 1 
#### Manejo de archivos y excepciones

Crear un programa en Python que:

- Pida al usuario un nombre de archivo.
- Intente abrir el archivo y leer su contenido línea por línea.
- Si el archivo no existe, capture la excepción y muestre el mensaje: "El archivo no fue encontrado".
- Si el archivo existe, cuente cuántas líneas contiene y muestre el total.
- Puntos extra: implementar una función genérica contar_lineas[T](archivo: T) que reciba el nombre del archivo como parámetro genérico.


````python
from typing import TypeVar

# Definimos un tipo genérico para mayor claridad (aunque no es estrictamente necesario)
T = TypeVar('T')

def contar_lineas(archivo: T) -> int:
    """
    Abre un archivo de texto y cuenta cuántas líneas contiene.
    Si el archivo no existe, maneja la excepción apropiadamente.
    """
    try:
        with open(archivo, 'r', encoding='utf-8') as f:
            lineas = f.readlines()
            return len(lineas)
    except FileNotFoundError:
        print("El archivo no fue encontrado.")
        return 0

def main():
    nombre_archivo = input("Ingrese el nombre del archivo: ")
    total = contar_lineas(nombre_archivo)
    if total > 0:
        print(f"El archivo tiene {total} líneas.")

if __name__ == "__main__":
    main()

# RESUMEN
"""
- Se usa with open(..., 'r') para abrir el archivo en modo lectura.
- readlines() devuelve una lista con todas las líneas.
- Si el archivo no existe, se captura la excepción FileNotFoundError.
- Se imprime un mensaje de error y se retorna 0.
- TypeVar('T') se incluye para ejemplificar el uso genérico (por ejemplo, si luego quisieras aceptar Path u otros tipos de ruta).
"""
````

### Ejercicio 2 
#### Decoradores y colecciones

Implementar un decorador llamado @registrar_llamada que:

- Imprima el nombre de la función y los argumentos con los que fue llamada.
- Luego ejecute la función normalmente.

Usar el decorador sobre una función buscar_producto(productos, nombre) que:
- Reciba una lista de diccionarios, donde cada diccionario representa un producto con las claves "nombre" y "precio".
- Busque un producto por su nombre y devuelva su precio, o None si no existe.

Ejemplo de uso:

```
productos = [
    {"nombre": "Mouse", "precio": 2000},
    {"nombre": "Teclado", "precio": 3500}
]

buscar_producto(productos, "Mouse")
```

Salida esperada:
```
Llamando a buscar_producto con argumentos: ([{'nombre': 'Mouse', 'precio': 2000}, {'nombre': 'Teclado', 'precio': 3500}], 'Mouse')
Precio encontrado: 2000
```

````python

# Decorador
def registrar_llamada(func):
    def wrapper(*args, **kwargs):
        print(f"Llamando a {func.__name__} con argumentos: {args}, {kwargs}")
        resultado = func(*args, **kwargs)
        return resultado
    return wrapper

# Función principal decorada
@registrar_llamada
def buscar_producto(productos, nombre):
    """
    Busca un producto por su nombre dentro de una lista de diccionarios.
    Devuelve el precio o None si no existe.
    """
    for producto in productos:
        if producto["nombre"].lower() == nombre.lower():
            print(f"Precio encontrado: {producto['precio']}")
            return producto["precio"]
    print("Producto no encontrado.")
    return None

if __name__ == "__main__":
    productos = [
        {"nombre": "Mouse", "precio": 2000},
        {"nombre": "Teclado", "precio": 3500},
        {"nombre": "Monitor", "precio": 9000}
    ]

    buscar_producto(productos, "Mouse")
    buscar_producto(productos, "Notebook")

# SALIDA ESPERADA
Llamando a buscar_producto con argumentos: ([{'nombre': 'Mouse', 'precio': 2000}, {'nombre': 'Teclado', 'precio': 3500}, {'nombre': 'Monitor', 'precio': 9000}], 'Mouse'), {}
Precio encontrado: 2000

Llamando a buscar_producto con argumentos: ([{'nombre': 'Mouse', 'precio': 2000}, {'nombre': 'Teclado', 'precio': 3500}, {'nombre': 'Monitor', 'precio': 9000}], 'Notebook'), {}
Producto no encontrado.

# RESUMEN
"""
- El decorador registrar_llamada recibe una función (func), define un wrapper que:
- Muestra el nombre de la función y sus argumentos.
- Ejecuta la función original.
- Devuelve su resultado.
- Se aplica con @registrar_llamada sobre la función buscar_producto.
- buscar_producto recorre la lista de diccionarios, compara el nombre (ignorando mayúsculas) y devuelve el precio si lo encuentra.
- Si no lo encuentra, devuelve None.
````
